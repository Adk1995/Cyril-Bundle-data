```python
import os
from PIL import Image
import pandas as pd
import tqdm
```

```python
products = {}
product_ID = []
base_url = 'data/300ppi/'
```

```python
df_BasicInfo = pd.read_csv('data/BasicProductDetails.csv')
df_BasicInfo['Company ID'] = df_BasicInfo['Company ID'].astype('str')
df_BasicInfo['Product ID'] = df_BasicInfo['Product ID'].astype('str').str.zfill(5)
df_BasicInfo.head()
```

```python
def create_product_dict(df_Info):
    products.clear()
    for index,company in enumerate(df_Info['Company ID']):
        if company in products.keys():
            products[company].append(df_Info['Product ID'][index])
        else:
            products[company] = [df_Info['Product ID'][index]]
```

```python
create_product_dict(df_BasicInfo)
products
```

```python
def fetch_images(product_dict):
    height = 0
    length = 0
    width =0
    dim = []
    dim1 = []
    image_exts = ['_1.png','_2.png','_3.png','_7.png','_8.png','_9.png']
    for key in tqdm.tqdm(product_dict.keys()):
        company_id = key
        for item in product_dict[company_id]:
            for ext in image_exts:
                if os.path.exists(base_url+company_id+'/'+item+ext):
                    image = Image.open(base_url+company_id+'/'+item+ext)
                    if '_1.png' in ext:
                        height = image.size[1]
                        width = image.size[0]
                        #print("Height = "+str(height))
                        #print("Width = "+str(width))
                        
                        if not os.path.exists(base_url+company_id+'/'+item+'_1.jpg'):
                            temp = image.convert('RGB')
                            temp.save(base_url+company_id+'/'+item+'_1.jpg')
                            
                    elif '_2.png' in ext:
                        length = image.size[0]
                        if not (image.size[1] == height and height>0):
                            dim.append(length)
                            dim.append(height)
                            image = image.resize(dim, Image.ANTIALIAS)
                        temp = image.convert('RGB')
                        temp.save(base_url+company_id+'/'+item+'_2.jpg')
                        del dim[:]
                    
                    elif '_3.png' in ext:
                        dim.append(width)
                        dim.append(length)
                        image.resize(dim, Image.ANTIALIAS)
                        temp = image.convert('RGB')
                        temp.save(base_url+company_id+'/'+item+'_3.jpg')
                        #print("Length = "+str(length))
                        del dim[:]
                        
                    elif '_7.png' in ext:
                        dim.append(width)
                        dim.append(height)
                        image.resize(dim,Image.ANTIALIAS)
                        temp = image.convert('RGB')
                        temp.save(base_url+company_id+'/'+item+'_7.jpg')
                        del dim[:]
                        
                    elif '_8.png' in ext:
                        dim.append(length)
                        dim.append(height)
                        image.resize(dim,Image.ANTIALIAS)
                        temp = image.convert('RGB')
                        temp.save(base_url+company_id+'/'+item+'_8.jpg')
                        del dim[:]
                    
                    elif '_9.png' in ext:
                        dim.append(width)
                        dim.append(length)
                        image.resize(dim,Image.ANTIALIAS)
                        temp = image.convert('RGB')
                        temp.save(base_url+company_id+'/'+item+'_9.jpg')
                        del dim[:]
```

```python
fetch_images(products)
```
